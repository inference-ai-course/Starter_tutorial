## Section 2: Data Structures

### Choosing the Right Data Structure

One of the most important skills in Python is selecting the appropriate data structure. The wrong choice can make code slow and complex; the right choice makes it elegant and fast.

**Decision Framework:**

| If you need...                          | Use this          | Because...                        |
|-----------------------------------------|-------------------|-----------------------------------|
| Ordered collection that changes         | **List**          | Flexible, mutable, O(1) append    |
| Fast lookup by unique key               | **Dictionary**    | O(1) access, perfect for mappings |
| Immutable, ordered sequence             | **Tuple**         | Memory efficient, hashable        |
| Unique items, set operations            | **Set**           | O(1) membership, deduplication    |

**Real AI/ML Examples:**

```python
# Lists: Training data batches
batch_data = [sample1, sample2, sample3, ...]

# Dicts: Model configurations
config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 100
}

# Tuples: Image dimensions (never change)
image_shape = (224, 224, 3)

# Sets: Unique vocabulary in NLP
vocabulary = {'the', 'cat', 'sat', 'on', 'mat'}
```

**Performance Tip:** Lists and tuples are contiguous in memory (fast iteration), while dicts and sets use hash tables (fast lookup).

## Section 3: Control Flow

### Understanding Loops and Conditionals

Control flow determines how your program executes. In AI/ML, you'll use these patterns constantly:

**Common Patterns:**

1. **Data Filtering** (if/else with comprehensions)
   ```python
   # Filter valid samples
   clean_data = [x for x in raw_data if validate(x)]
   ```

2. **Batch Processing** (for loops)
   ```python
   # Process data in batches
   for batch in dataloader:
       predictions = model(batch)
   ```

3. **Training Loops** (nested loops with conditions)
   ```python
   for epoch in range(num_epochs):
       for batch in train_data:
           if batch_idx % 100 == 0:
               evaluate_model()
   ```

4. **Early Stopping** (while with break)
   ```python
   while not converged:
       loss = train_step()
       if loss < threshold:
           break
   ```

**Performance Consideration:**
- List comprehensions are faster than explicit loops
- Use `enumerate()` instead of manual counters
- `break` and `continue` make code cleaner than complex conditionals

## Section 4: Functions

### The Building Blocks of Organized Code

Functions transform spaghetti code into organized, testable systems. This is especially critical in AI/ML where pipelines have many stages.

**Why Functions Matter:**

```python
# âŒ Bad: Everything in one place (script mentality)

def load_data(path):
    return pd.read_csv(path)

def clean_data(df):
    return df.dropna()

def normalize_features(df, column):
    return (df[column] - df[column].mean()) / df[column].std()

def train_model(X, y):
    model = RandomForest()
    model.fit(X, y)
    return model

def evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    return (predictions == y_test).mean()

# Now each piece is testable, reusable, and clear!
```

**Function Design Principles:**

1. **Single Responsibility**: One function = one job
2. **Clear Naming**: Function name describes what it does
3. **Type Hints**: Document expected types
4. **Docstrings**: Explain purpose, args, and returns
5. **Small Size**: If it doesn't fit on one screen, break it up

**In AI/ML Pipelines:**
- Data loading â†’ function
- Preprocessing â†’ function
- Feature engineering â†’ function
- Model training â†’ function
- Evaluation â†’ function

This makes experimentation easier (swap functions) and debugging faster (isolate problems).

### Exception Handling: Production-Ready Code

Errors will happen. The question is: Does your code crash or recover gracefully?

**Exception Handling Mindset:**

```python
# Student code (crashes on error):
data = json.loads(file_content)
model.fit(data)

# Production code (handles errors):
try:
    data = json.loads(file_content)
    validate_data_schema(data)
    model.fit(data)
except json.JSONDecodeError as e:
    logger.error(f"Invalid JSON: {e}")
    return use_default_config()
except ValidationError as e:
    logger.warning(f"Bad data: {e}")
    return use_fallback_data()
except Exception as e:
    logger.critical(f"Unexpected error: {e}")
    raise
```

**Why This Matters in AI/ML:**
- Training runs for hours â†’ need checkpointing on errors
- Data pipelines process millions of files â†’ some will be corrupt
- API calls (OpenAI, HuggingFace) â†’ networks fail
- **Robust error handling prevents data loss and wasted compute time**
